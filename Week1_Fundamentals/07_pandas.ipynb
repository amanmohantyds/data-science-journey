### i) Series

### Example 1

```python
data_mark=[0, 35, 56, 98, 55]
series=pd.Series(data_mark)
print(series)
```

```python
#making a series with custom index
data_mark=[0, 35, 56, 98, 55]
series_2=pd.Series(data_mark, index=['a', 'b', 'c', 'd', 'e'])
print(series_2)
```

```python
#Accessing by label
data_mark=[0, 35, 56, 98, 55]
print(series_2.loc["c"])
```

```python
#Accessing by index
data_mark=[0, 35, 56, 98, 55]
print(series_2.iloc[2])
```

```python
#Changing the value of a specific index using a label
data_mark=[0, 35, 56, 98, 55]
series_2.loc["c"] =12
print(series_2)
```

### Data Frame

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])
print(df)
```

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])
print(df.loc["Employee2"])
#Accessing a row by index label
```

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
    
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])
print(df.iloc[1])
#Accessing a row by index label 
```

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])

#Add a new column
df["Company"] = ["TCS", "Infosys", "Wipro"]
print(df)  #Fortune-Telling Lady
```

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])
df["Company"] = ["Crusty Crab", "Crusty Crab", "N/A"]
#Add a new row
new_row= pd.DataFrame([{"Name": "Sandy", "Age": 16, "Roll No.": 75, "Company": "N/A"},
                       {"Name": "Mr. Crab", "Age": 55, "Roll No.": 5, "Company": "CEO"}],
                    index=["Employee4", "Employee5"])
df= pd.concat([df, new_row])
print(df)
```

### Importing CSV and JSON Files

```python
#Importing  a csv file
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")
print(df)
```

```python
#To get full file 
df=pd.read_csv("pokemon_dataset.csv")
print(df.to_string())
```

```python
# selection by columns
df=pd.read_csv("pokemon_dataset.csv")
print(df["Name"].to_string())
```

```python
df=pd.read_csv("pokemon_dataset.csv")
print(df[["Name", "Type 1", "Type 2"]].to_string())
```

```python
#selection by row/s
df=pd.read_csv("pokemon_dataset.csv")
print(df.loc[9])
```

```python
# selection using name
df=pd.read_csv("pokemon_dataset.csv", index_col="Name")
print(df.loc["Pikachu"])
```

```python
#Access specific column(i.e. "Type 1", "Type 2", "HP") of specific row(i.e. "Pikachu")
df=pd.read_csv("pokemon_dataset.csv", index_col="Name")
print(df.loc["Pikachu", ["Type 1", "Type 2", "HP"]])
```

```python
#Use Slicing in loc to access multiple rows and columns
df=pd.read_csv("pokemon_dataset.csv", index_col="Name")
print(df.loc["Pikachu":"Meowth", ["Type 1", "Type 2", "HP"]])
```

```python
#Using iloc to access specific rows and columns by index
df=pd.read_csv("pokemon_dataset.csv")
print(df.iloc[0:7, [1, 2, 3]])
```

```python
#Slicing and accessing every second column
df=pd.read_csv("pokemon_dataset.csv")
print(df.iloc[0:7:2, [1, 2, 3]])
```
