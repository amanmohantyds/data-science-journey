Using .iloc[ ] function for Accessing value using specific Row number and Column number

```python
import pandas as pd
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65]
    
}
df=pd.DataFrame(data)


# Access the value on the first row and the first column
print(df.iloc[0, 0])

# Access the value on the first row and the third column
print(df.iloc[0,2])
```

Adding a New Column

```python
import pandas as pd
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55,65]
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])

#Add a new column
df["Company"] = ["TCS", "Infosys", "Wipro"]
print(df)
```

Using .concat() function to add a new directory/tuple to existing directory/tuple

```python
data= {
    "Name" : ["SpongeBob","Patrick","Squidward"],
    "Age" : [17, 19, 18],
    "Roll No." : [45, 55, 65] 
}
df=pd.DataFrame(data, index=["Employee1", "Employee2", "Employee3"])
df["Company"] = ["Crusty Crab", "Crusty Crab", "N/A"]
#Add a new row
new_row= pd.DataFrame([{"Name": "Sandy", "Age": 16, "Roll No.": 75, "Company": "N/A"},
                       {"Name": "Mr. Crab", "Age": 55, "Roll No.": 5, "Company": "CEO"}],
                    index=["Employee4", "Employee5"])
df= pd.concat([df, new_row])
print(df)
```

To Read data from Excel

```python
from pyodide.http import pyfetch
import pandas as pd
async def download(url, filename):
    response = await pyfetch(url)
    if response.status == 200:
        with open(filename, "wb") as f:
            f.write(await response.bytes())
            
xlsx_path = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n9LOuKI9SlUa1b5zkaCMeg/Product-sales.xlsx'

await download(xlsx_path, "Product-sales.xlsx")

df = pd.read_excel("Product-sales.xlsx")
print(df.head())
```

### Data Cleaning

- It is process of fixing/removing :
   incomplete, incorrect, or irrelevant data.
- ~75% of work done with Pandas is data cleaning    

#### i ) Drop Relevant Columns

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")

#drop irrelevant columns
df = df.drop(columns=["Total", "Generation"])
print(df)
```

#### ii ) Handle Missing Data

##### a ) Drop Missing values

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")


df = df.dropna(subset=["Type 2"])
print(df.to_string())
```

##### b ) Fill Missing Values

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")


df=df.fillna({"Type 2": "None"})
print(df.to_string())
```

##### c ) Fill Any Inconsistent Values

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")

df["Type 1"] = df["Type 1"].replace({"Fire": "Fire Type",
                                     "Water": "Water Type",
                                     "Grass": "Grass Type"})
print(df.to_string())
```

##### d ) Standardize Text

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")

df["Generation"] = df["Generation"].astype(bool)
print(df.to_string())
```

##### e ) Remove Duplicate Values

```python
import pandas as pd
df=pd.read_csv("pokemon_dataset.csv")

df=df.drop_duplicates()

print(df.to_string())
```
