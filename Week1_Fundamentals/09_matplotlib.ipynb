### Plt Customization

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y = np.array([15, 25, 30, 20])

plt.plot(x, y, marker=".",
               markersize=15,
               markerfacecolor="#1cd3fc",
               markeredgecolor="#1cd3fc",\
               linestyle="dashdot",
               linewidth=1,
               color="#1c5bfc")

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

line_style = dict(marker=".",
               markersize=15,
               markerfacecolor="#1cd3fc",
               markeredgecolor="#1cd3fc",
               linestyle="solid",
               linewidth=2
                                     )

plt.plot(x, y1, color="#1c5bfc", **line_style)

plt.plot(x, y2,color="#1cfc45", **line_style)

plt.plot(x, y3, color="#fc1c1c", **line_style)

plt.show()
```

### Labels

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])


plt.title("Class Size", fontsize=25,
                        family="Arial",
                        fontweight="bold",
                        color="#2d4cfc")

plt.xlabel("Year", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#2dbefc")

plt.ylabel("Students", fontsize=20,
                   family="Arial",
                   fontweight="bold",
                   color="#2dbefc")

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.show()
```

To add ticks at specific places or point on x axis where we write points or values
- we can make it to generate ticks at given points we want

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)

plt.xticks(x)

plt.show()
```

To Customize ticks use 'plt.ticks_params(axis=" ", color= " ")

```python
import matplotlib.pyplot as plt 
import numpy as np

x = np.array([2023, 2024, 2025, 2026])
y1 = np.array([15, 25, 30, 20])
y2 = np.array([17, 23, 38, 5])
y3 = np.array([13, 15, 20, 30])

plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)


plt.xticks(x)
plt.tick_params(axis="both", color="#2dbefc")

plt.show()
```

### Grid

For a single axis grid

```python
import matplotlib.pyplot as plt 
import numpy as np

x = [1, 2, 3, 4, 5]
y = [5, 10, 15, 20, 25]

plt.grid(axis="y",
         linewidth=2,
         color="lightgrey",
         linestyle="dashed")

plt.plot(x, y)
plt.show()
```

### Pie Charts

```python
import matplotlib.pyplot as plt 
import numpy as np

categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

plt.pie(values, labels=categories)

plt.show()
```

To add percentages in pie charts

```python
import matplotlib.pyplot as plt 
import numpy as np

categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

plt.pie(values, labels=categories,
                autopct="%1.2f%%")    
# To add how many decimal points should be there in percentage then add a point after 1 and add the number of decimal places
# To add a percentage sign add %%  to "autopct="%1.1f" " --> autopct="%1.1f%%"

plt.show()
```

To add color in the pie charts

```python
import matplotlib.pyplot as plt 
import numpy as np

categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

colors= ["red", "yellow", "blue", "green"]

plt.pie(values, labels=categories,
                autopct="%1.2f%%",
                colors=colors,
                )    

plt.show()
```

To explode a pie slice and drop its shadow

```python
import matplotlib.pyplot as plt 
import numpy as np

categories = ["Freshman", "Sophomores", "Juniors", "Seniors"]
values = np.array([300, 250, 275, 225])

plt.pie(values, labels=categories,
                autopct="%1.2f%%",
                colors=colors,
                explode=[0, 0, 0, 0.1],    #To explode
                shadow=True,     # To add shadow
                startangle=90      # To rotate the chart
                )    


plt.show()
```

### Scatter Chart

```python
import matplotlib.pyplot as plt 

x1 = [0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8]
y1 = [55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87]

plt.scatter(x1, y1, color="Skyblue",   #Color
                  alpha= 0.5,    #transparency of dots
                  s = 100)     # size of dots

plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.show()
```

```python
import matplotlib.pyplot as plt 
import numpy as np

x1 = np.array([0, 1, 1, 2, 3, 4, 5, 6, 7, 7, 8])
y1 = np.array([55, 60, 65, 62, 68, 70, 75, 78, 82, 85, 87])

x2 = np.array([0, 1, 2, 2, 3, 4, 5, 6, 7, 8, 8])
y2 = np.array([50, 58, 65, 70, 72, 78, 83, 88, 92, 95, 97])

plt.scatter(x1, y1, color="skyblue",
                    alpha= 0.5,  
                    s = 100,
                    label="Class A") 

plt.scatter(x2, y2, color="red",
                    alpha= 0.5,  
                    s = 100,
                    label="Class B") 


plt.title("Test Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Grade")

plt.legend()     # To show labels  in graph use plt.legend()

plt.show()
```

### Histograms

- A visual representation of the distributions of quantitative data.
- They group values into bins (intervals) and counts how many fall in each range

```python
import matplotlib.pyplot as plt 
import numpy as np

# To are some keyword arguments to pass like "loc" which location of centre or median of our data, "scale" is standard deviation or spread ie how far it is going to deviate ,
# so higher the number higher the deviation and "size" is how many numbers we want to generate

scores = np.random.normal(loc=80, scale=10, size=100)

# Just to ensure no value goes below zero or above 100, we can use clip function
scores = np.clip(scores, 0, 100)

plt.hist(scores, bins=10,
                 color="lightgreen",
                 edgecolor="black")

# We can set the number of bins or intervals we want using 'bins' keyword argument,
# plt.hist(scores, bins=10)

plt.title("Exam Scores")
plt.xlabel("Score")
plt.ylabel("# of students")

plt.show()

```

### Subplots

- Figure = The entire canvas 
- Ax = A single plot (subplot)

```python
import matplotlib.pyplot as plt 
import numpy as np

x=np.array([1, 2, 3, 4, 5])


#  plt.subplot(rows, columns)
figure, axis = plt.subplots(2, 2)

axis[0, 0].plot(x, x*2, color="red")
axis[0, 0].set_title("x*2")

axis[0, 1].plot(x, x**2, color="blue")
axis[0, 1].set_title("x**2")

axis[1, 0].plot(x, x**3, color="green")
axis[1, 0].set_title("x**3")

axis[1, 1].plot(x, x**4, color="purple")
axis[1, 1].set_title("x**4")

# To avoid overlapping of graph use 'tight_layout()'
plt.tight_layout()

plt.show()
```
