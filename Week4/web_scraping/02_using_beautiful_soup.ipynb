# Web Scrapping

#### 1) Using local html files

```python
from bs4 import BeautifulSoup

with open("Week1_Fundamentals/12.1_file.html", "r", encoding="utf-8") as html_file:
    content = html_file.read()
    
    soup = BeautifulSoup(content, "lxml")
    source_cards = soup.find_all("div", class_="container")
    for source in source_cards:
       print(source.text)
```
